```python
# Following video from 6.4.1
from scipy.stats import linregress
import matplotlib.pyplot as plt
```

```python
temps = [18.2, 29.5, 35.9, 50.1, 58.6, 69.6, 82.1, 85.5, 73.5, 55.2, 40.3, 30.1]
daylight = [7.6, 9.2, 12.2, 13.4, 13.9, 14.2, 13.5, 11.8, 10.9, 9.9, 8.4, 7.8]
print(len(temps))
print(len(daylight))
```

```python
plt.scatter(daylight, temps)
plt.xlabel('hours of daylight')
plt.ylabel('average monthly temp')
plt.show()
```

```python
linregress(daylight, temps)
```

```python
# save results as a tuple
(slope, intercept, rval, pval, stderr) = linregress(daylight,temps)
```

```python
# Create line equation
line_eq = "y = " + str(round(slope, 2)) + "x + " + str(round(intercept,2))
print(line_eq)
                                                    
```

```python
regress_value = [(day * slope + intercept) for day in daylight]
plt.scatter(daylight, temps)
plt.plot(daylight, regress_value)
plt.xlabel('hours of daylight')
plt.ylabel('average monthly temp')
plt.show()
```
